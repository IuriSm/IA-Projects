Criando os dataframes das reviews positivas e negativas e adicionado o tipo de classse de em cada uma delas.




```python
datap_df = pd.DataFrame(datap,columns=['Texto'])
datan_df = pd.DataFrame(datan,columns=['Texto'])
datan_df["Classe"] = "Negativo"
datap_df["Classe"] = "Positvo"
```

```python
datap_df
```

Juntado ambos os dataframes das reviews positivas e negativas em uma coluna só e dropando as anteriores.

Embaralhando o dataframe para as linhas não ficarem divididas em um ponto específico.

Passando os dados de object para string.

```python
danp_df = datap_df.append(datan_df).dropna()
danp_df = danp_df.sample(frac=1).reset_index(drop=True)
danp_df = danp_df.astype({ 'Texto' : 'string'})
danp_df = danp_df.astype({ 'Classe' : 'string'})
```

```python
danp_df.info()
```

```python
danp_df
```

Verificando se se existe linhas únicas e duplicadas.
Em classe contém linhas duplicadas e sem ser únicas, e nos textos contém linhas únicas e sem duplicatas.

```python
clone = danp_df['Classe'].is_unique
clones = danp_df['Classe'].duplicated().any()
print(clone,"e",clones)
```

```python
clone = danp_df['Texto'].is_unique
clones = danp_df['Texto'].duplicated().any()
print(clone,"e",clones)
```

Um exemplo dos tipos de procura do re.

```python
p1 = re.compile(r'\W', re.A)
p1.search(review)
```

Vendo caracteres que não letras.

```python
p2 = r'\W'
re.findall(p2, review)
```

Um dos modos de separação por espaço.

```python
p3 = r'(?<![\"=\w])(?:[^\W_]+)(?![\"=\w])'
re.findall(p3, review)
```

Outro modo parecido mas com diferenças em algumas escolhas como "pretão" que é considerado palavra em um e não em outro.

```python
p4 = r'[\w]+|[a-zA-Z0-9_]'
lista = re.findall(p4, review)
lista
```

Separando especificamente pelo espaço entre os caracteres e pegando todo o resto.

```python
lista = re.split(' +', review)
lista
```
