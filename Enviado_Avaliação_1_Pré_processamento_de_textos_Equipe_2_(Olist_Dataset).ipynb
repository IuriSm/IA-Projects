Criando os dataframes das reviews positivas e negativas e adicionado o tipo de classse de em cada uma delas.




```python
datap_df = pd.DataFrame(datap,columns=['Texto'])
datan_df = pd.DataFrame(datan,columns=['Texto'])
datan_df["Classe"] = "Negativo"
datap_df["Classe"] = "Positvo"
```

```python
datap_df
```

Juntado ambos os dataframes das reviews positivas e negativas em uma coluna s√≥ e dropando as anteriores.

Embaralhando o dataframe para as linhas n√£o ficarem divididas em um ponto espec√≠fico.

Passando os dados de object para string.

```python
danp_df = datap_df.append(datan_df).dropna()
danp_df = danp_df.sample(frac=1).reset_index(drop=True)
danp_df = danp_df.astype({ 'Texto' : 'string'})
danp_df = danp_df.astype({ 'Classe' : 'string'})
```

```python
danp_df.info()
```

```python
danp_df
```

Verificando se se existe linhas √∫nicas e duplicadas.
Em classe cont√©m linhas duplicadas e sem ser √∫nicas, e nos textos cont√©m linhas √∫nicas e sem duplicatas.

```python
clone = danp_df['Classe'].is_unique
clones = danp_df['Classe'].duplicated().any()
print(clone,"e",clones)
```

```python
clone = danp_df['Texto'].is_unique
clones = danp_df['Texto'].duplicated().any()
print(clone,"e",clones)
```
